```python
import numpy as np

arr = np.array([1, 3, 5, 7])
print('numpy_array:', arr)

arr.shape
```

```python
arr2 = np.array([[1, 3, 5, 7], [4, 3, 5, 9]])
print('numpy_array2:', arr2)

arr2.shape
```

```python
#dtype
a = np.array([1, 2, 3])
print(a.dtype)
```

```python
# reshape ()
a = np.array([1, 2, 3, 4, 6, 7])
b = a.reshape(2,3)
print('shaped_array:', b)
```

```python
# astype()
a = np.array([1, 2, 3])
b = a.astype(float)
print('float:', b)
```

```python
# np.zeros(), np.ones() => dummy data
import numpy as np
z = np.zeros((2, 3))
o = np.ones((2, 3))

print("zeros:", z)
print('ones:', o)
```

```python
# ndim (nechta array borligini)
a = np.array([[1, 2], [3, 4]])
print('ndim:', a.ndim)
```

```python
# arange(oxirgi sonni o'zini chiqarmaydi)
a = np.arange(1, 6)
print('arange:', a)
```

```python
# sum(), mean() - sonlarni o'rtacha, max(), min
```

```python
# flatten()
a = np.array([[1, 2], [3, 4]])
print('flatten:', a.flatten())
```

```python
# T - transpose
a = np.array([[1, 2], [3, 4]])
print('flatten:', a.T)
```

```python
# dot() @ -- ko'paytirish
a = np.array([[1, 2], [3, 4]])
b = np.array([[2, 0], [1, 2]])
print('dot:', a.dot(b))
```

```python
# random.rand() -- random sonlar => 0.0 ~ 1.0 (1.0 excluded)
a = np.random.rand(2, 3)
print('random rand:', a)
```
